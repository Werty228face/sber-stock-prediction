```python
!pip install -q yfinance

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# 1. Скачиваем данные (акции Сбера)
print("Загружаем данные...")
df = yf.download('SBER.ME', start='2020-01-01', end='2025-01-01', progress=False)

# Проверяем
print(f"Загружено строк: {len(df)}")
print(df.head())

# 2. Берём только цену закрытия
data = df[['Close']].copy()
dataset = data.values

print(f"Размер массива: {dataset.shape}")

# 3. Масштабируем
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

print("Масштабирование выполнено!")

# 4. Строим график цены
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Close'], label='Цена закрытия Сбера', color='blue')
plt.title('Динамика акций Сбербанка')
plt.xlabel('Дата')
plt.ylabel('Цена (руб)')
plt.legend()
plt.grid()
plt.show()
```

```python

# 1. Создаем датасет для обучения (с окном в 60 дней)
training_data_len = int(np.ceil(len(scaled_data) * 0.95))  # 95% на обучение

train_data = scaled_data[0:training_data_len, :]

# Создаем x_train и y_train
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])

# Превращаем в numpy массивы
x_train, y_train = np.array(x_train), np.array(y_train)

# Для LSTM нужно 3D: [samples, time_steps, features]
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

print(f"x_train shape: {x_train.shape}")  # (кол-во, 60, 1)
print(f"y_train shape: {y_train.shape}")

# 2. Строим модель LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(60, 1)))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dropout(0.2))
model.add(Dense(1))

# Компилируем
model.compile(optimizer='adam', loss='mean_squared_error')

# 3. Обучаем модель
print("\nНачинаем обучение...")
history = model.fit(x_train, y_train, epochs=20, batch_size=32, verbose=1)

print("Обучение завершено!")

# 4. Тестируем на оставшихся данных (5%)
test_data = scaled_data[training_data_len - 60:, :]

x_test = []
y_test = scaled_data[training_data_len:, :]

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

# Предсказываем
predictions = model.predict(x_test)

# Возвращаем в исходный масштаб
predictions = scaler.inverse_transform(predictions)
y_test_inv = scaler.inverse_transform(y_test)

# 5. Визуализируем результат
plt.figure(figsize=(12, 6))
plt.plot(y_test_inv, label='Реальная цена', color='blue')
plt.plot(predictions, label='Предсказанная цена', color='red')
plt.title('Предсказание акций Сбера (LSTM)')
plt.xlabel('Дни (5% тестовой выборки)')
plt.ylabel('Цена (руб)')
plt.legend()
plt.grid()
plt.show()

# 6. Считаем ошибку
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

mae = mean_absolute_error(y_test_inv, predictions)
mse = mean_squared_error(y_test_inv, predictions)
rmse = math.sqrt(mse)

print(f"\n📊 Оценка качества модели:")
print(f"Средняя абсолютная ошибка (MAE): {mae:.2f} руб.")
print(f"Среднеквадратичная ошибка (RMSE): {rmse:.2f} руб.")
```
